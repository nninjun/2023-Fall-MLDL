### Q5(e)

#### You can see the plots below highlighting the relationships among the predictors.  If an absolute value of correlation coefficient is bigger than 0.8, than we can say two variables are strongly related.

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

scatter_matrix(quant_data, alpha=0.5,figsize=(10,10))
plt.show()
```

#### I made additional plot using module "seaborn"

```python
import seaborn as sns

sns.set(font_scale=1.0)
sns.set_style('ticks')
sns.pairplot(quant_data, diag_kind=None) #if we set diag_kind='Hist', then we get histograms for diagonal graphs like those above(scatter matrix of pandas).
plt.show()
```

#### Through quant_data.corr(), we knew that mpg has strong correlation with weight, displacement, horsepower, cylinders

```python
fig, axes = plt.subplots(ncols=4, figsize=(16, 4))
quant_data.plot.scatter('weight', 'mpg', ax=axes[0],color='red')
quant_data.plot.scatter('displacement', 'mpg', ax=axes[1], color='blue')
quant_data.plot.scatter('horsepower', 'mpg', ax=axes[2], color='green')
quant_data.plot.scatter('cylinders', 'mpg', ax=axes[3], color='yellow')
axes[0].set_title("weight vs mpg")
axes[1].set_title("displacement vs mpg")
axes[2].set_title("horsepower vs mpg")
axes[3].set_title("cylinders vs mpg")
```

#### mpg also has some correlation with year and acceleration, but it's small compared to above

```python
fig, axes = plt.subplots(ncols=2, figsize=(8, 4))
quant_data.plot.scatter('year', 'mpg', ax=axes[0],color='red')
quant_data.plot.scatter('acceleration', 'mpg', ax=axes[1], color='blue')
axes[0].set_title("year vs mpg")
axes[1].set_title("acceleration vs mpg")
```
