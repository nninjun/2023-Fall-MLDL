```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
user_profile = pd.read_csv('./UserPorfile.csv')
submission_record = pd.read_csv('./SubmissionRecord.csv')
```

```python
#I wanted to see data distributions in UserProfile
sns.set(style="whitegrid")
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

#Status Distribution
sns.countplot(x='Status', data=user_profile, ax=axes[0, 0])
axes[0, 0].set_title('Status Distribution')

#Year Distribution
sns.countplot(x='Year', data=user_profile, ax=axes[0, 1])
axes[0, 1].set_title('Year Distribution')

#Department Distribution
sns.countplot(x='Department', data=user_profile, ax=axes[0, 2])
axes[0, 2].set_title('Department Distribution')
axes[0, 2].tick_params(axis='x', rotation=45)

#Credits Distribution
sns.histplot(user_profile['# Credits'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Credits Distribution')

#GPA Distribution
sns.histplot(user_profile['GPA'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('GPA Distribution')

#IQ Distribution
sns.histplot(user_profile['IQ'], kde=True, ax=axes[1, 2])
axes[1, 2].set_title('IQ Distribution')

#Study Hour Distribution
sns.histplot(user_profile['Study Hour'], kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Study Hour Distribution')

#MBTI Distribution
sns.countplot(x='MBTI', data=user_profile, ax=axes[2, 1])
axes[2, 1].set_title('MBTI Distribution')
axes[2, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
axes[2, 2].axis('off')
plt.show()
```
